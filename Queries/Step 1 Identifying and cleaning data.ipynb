```python
import re
```

```python
def extract_brackets(text):
    return re.findall(r'\((.*?)\)', text)
```

```python
df['Tags']=df['skills'].apply(extract_brackets)
```

```python
df['Tags'].head(20)
```

```python
def remove_eg(tags):
    cleaned_tags = []
    for tag_list in tags:
        if isinstance(tag_list, str):
            # Remove 'e.g.,' from the tag
            cleaned_tag = re.sub(r'e\.g\.,\s*', '', tag_list)
            # Split by commas and strip extra spaces
            cleaned_tags.extend([item.strip() for item in cleaned_tag.split(',') if item.strip()])
    return cleaned_tags
```

```python
df['Tags']=df['Tags'].apply(remove_eg)
```
